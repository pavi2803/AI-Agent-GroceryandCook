```python
import re

# Basic stopwords list (can expand later)
stopwords = set([
    "to", "for", "and", "or", "cut", "into", "cups", "cup", "tablespoons", "tablespoon",
    "divided", "pounds", "pound", "stick", "sticks", "teaspoons", "teaspoon", "tbsp", "tsp",
    "unsalted", "plus", "large", "small", "medium", "sliced", "chopped", "minced", "optional",
    "pieces", "from", "about", "in", "on", "of", "the", "a", "with", "casings", "removed",
    "lightly", "beaten", "coarsely", "fresh", "extra", "low", "high", "fat", "cut", "qt", "oz", "lb", "lbs", "g", "kg", "ml", "clove", "cloves"
])

def clean_ingredient_string(ingredient_str):
    # Lowercase and remove punctuation
    ingredient_str = re.sub(r"[^\w\s]", "", ingredient_str.lower())
    tokens = ingredient_str.split()
    # Filter out stopwords and short junk
    filtered = [word for word in tokens if word not in stopwords and len(word) > 2]
    # Deduplicate and join
    deduped = list(dict.fromkeys(filtered))  # preserves order
    return ", ".join(deduped)


```
